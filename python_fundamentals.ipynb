## **and** - Logical Operator

- It evalutes from left to right.
- It evaluates all expressions and returns the last expression, if all expressions evaluates to True.
- Otherwise, it returns first false value that evaluates to False.

```python
a = 10
b = 15

print(a and b)
print(b and a) # 10 - return last true value

# Tried with falsey values
my_list = []
my_set = {}
my_tuple = ()

print(my_list and my_set and my_tuple) # [] - return first false value
print(a and my_set and my_tuple) # {} - return first false value
print(None and None)
```

## **or** - Logical Operator

- It also evalutes from left to right.
- It returns the first value that evaluates set to True.
- Otherwise, it returns last value that evaluates set to False.

```python
a = 10
b = 15

print(a or b)
print(b or a) # 15 - return first true value

# Tried with falsey values
my_list = []
my_set = {}
my_tuple = ()

print(my_list or my_set or my_tuple) # () - return last false value
print(my_set or a or my_tuple) # 10 - return first true value
print(None or None)
print(0 or (10-10) or my_list or "")
print(repr(0 or (10-10) or my_list or ""))
```

## **not** - Logical Operator
- It returns boolean value only.

```python
a = 10
b = 20
c = 'Vignesh'

print(a == 10)
print(a == 20)
print(a == 'Vignesh')

# throws error
# print(a is 10)
# print(a is 20)
# print(a is "Vignesh")

print(a is a)
print(a is b)
print(a is c)
```

```python
a = 10
b = 20
c = 'Vignesh'

print(a is not a)
print(a is not b)
print(a is not c)
```

## docstring
- Used for documenting about a function that defines inside a function. It defines like `multiline comments`. But, if we declare at the nextline of function name, it called as docstring.
- A function can have only one docstring, at the same time can have multiple multiline comments. first one will be consider as docstring.
- To read the docstring, use `__doc__` magic/special method.
- If function contains __ as prefix and suffix, that is called as magic method.
- Please check the example below.


```python
def myFullName(firstName : str, lastName : str, age : int):
  '''
  This function represents display full name of the user
  @firstname receives the first name of the user
  @lastname receives the last name of the user
  '''
  '''
  Hi,
    above one is doc string, this is multiline comment
  '''

  return firstName + ' ' + lastName

print(myFullName('Vignesh', 'Murugesan', 2 ))

# to see the docstring of the function
print(myFullName.__doc__)
```

## Casting

- Convert a value/variable to another one data type.
- There are two types of Type Casting.
  - Implicit Type Casting
  - Explicit Type Casting

### Implicit Type Casting
- Converts a data type into another data type automatically.

```python
a = 10
b = 0.1

c = a * b

print(c)
print(type(c))
```

### Explicit Type Casting
- Here type casting with user involvement.

```python
my_str = '20'
print(type(my_str))

# string to interger
my_str = int(my_str) + 10
print(type(my_str))

my_str = float(my_str)
print(my_str)
print(type(my_str))

my_str = str(my_str)
print(my_str)
print(type(my_str))
```

## F-String

- Used for simplifying string formatting.
- Used to embed expressions and variables directly into a string.
- A f-string is having prefix with letter **f**.
- It works same as `str.format()`

### F-string with variables

```python
name = "Vignesh"
company = "Mitrahsoft"

# simple string
greet = "Hello, " + name + "\nWelcome to Mitrahsoft"
print(greet)

# f-string
fGreet = f"Hello, {name} \nWelcome to Mitrahsoft"
print(fGreet)

# f-string with multiple variable
fGreet = f"Hello, {name} \nWelcome to {company}"
print(fGreet)
```

### F-string with expressions

```python
a = int(input("Enter value to square: "))
print(f"Square value of given number is {a*a}")
```

### Using f-string to print date as well

```python
from datetime import datetime

print(f"Today is {datetime.today():%B, %d, %Y}")
```

## Loops

* A loop is a way to run the same block of code multiple times without writing it repeatedly.
* Common types: **for** loop (runs for a fixed number of times or over items) and **while** loop (runs until a condition becomes false).
* Loops help automate repetitive tasks, like processing lists, reading files, or performing calculations repeatedly.


### For loop

* A **for loop** is used to repeat code for each item in a sequence (like a list, string, or range of numbers).
* It automatically moves to the next item after each iteration.

```python
for i in range(5):
      print(i)  # prints 0 to 4
```

### While loop

* A **while loop** runs as long as a given condition is `True`.
* It checks the condition before each iteration and stops when it becomes `False`.

```python
count = 0
while count < 5:
    print(count)
    count += 1
```
