**File Modes:**
  * Once we perform any operation, it is mandatory to mention their mode like(read,write,append etc)
  * The file operation will be based on the file mode specified in it
  



**Mostly Used file modes in python:**
  * **r** - Read mode is used to read the file , you can able to read the whole content written in the file
  * **w** - Opens the file for writing. Creates a new file or truncates the existing file.
  * **a** - Opens the file for appending data. Creates a new file if it doesnt exist.
  * **x** - Creates a new file. Raises an error if the file already exists.
  * **r+** - Opens the file for both reading and writing. File must exist; otherwise, it raises an error.
  * **w+** - Opens the file for both writing and reading. Creates a new file or truncates the existing file.
  * **a+** - Opens the file for appending and reading. Creates a new file if it doesnt exist.
  * **x+** - Creates a new file for reading and writing. Raises an error if the file exists

**File Write:**
  * **Syntax:** file_name.write(content)

```python
file_write = open('testsample.txt','w')
file_write.write("hii this is testing")
# file_write.close()
```

**Writelines:**
  * The writelines() method is used to write multiple strings to a file.
  * The writelines() method takes an iterable object like lists, tuple, etc. containing strings as an argument.
  * **Syntax:**file_name.writelines(content)

```python
file_write = open('test.txt','w')
list_of_string = ["navis","\t smith"]
file_write.writelines(list_of_string)
file_write.close()
```

**File Read():**
  * **synax:** file_name.read(size)
  * Size:if you define the size, it reads the given size of characters in the file
  * readline is used to read the specific line not all line

```python
file_read = open('test.txt','r')
readed_content = file_read.read()
print(readed_content)
file_read.close()
```

**Readline and readlines:**
  * **Readline:**
    * The readline() method returns one line from the file
    * You can also specified how many bytes from the line to return, by using the size parameter
    * **Syntax:** file.readline(size)
  * Readlines:
    * The readlines() method returns a list containing each line in the file as a list item
    * Use the hint parameter to limit the number of lines returned. If the total number of bytes returned exceeds the specified number, no more lines are returned.

```python
file_read = open('testsample.txt','r')
readed_content = file_read.read()
# readed_content = file_read.read(5)
# readed_content = file_read.readlines()
print(readed_content)
print(file_read.tell())
# file_read.close()
```

**FIle Appending:**
  * Append means to add something to the end of an existing item, list, or file without changing the original content.
  * For example, when you append text to a file, it adds the new text at the end of the file without removing or altering what's already there.






```python
append_file = open('testsample.txt','a')
append_file.write("\ncontent are stored here")
append_file.close()
```

**File Seek():**
  * The file.seek() method is used to change the file's current position or "cursor" in a file.
  * It allows you to move to a specific byte in the file, so that you can read from or write to a particular location. (i.e., reading from or writing to arbitrary locations in the file).
  * **Syntax:** file.seek(offset, whence)
  

```python
combine_operation = open("testingwork.txt",'a+')
combine_operation.write("The combined operation works properly")
# combine_operation.seek(0)
combine_operation_words = combine_operation.read()
print(combine_operation_words)
```

**File tell():**
  * The Python File tell() method is used to find the current position of the file cursor (or pointer) within the file.
  * This method is mostly used in scenarios where there is a need to determine whether the file cursor is either at the beginning of the file or the ending.

```python
file_read = open('testsample.txt','r')
print(file_read.tell())
readed_content = file_read.read()
print(readed_content)
```
