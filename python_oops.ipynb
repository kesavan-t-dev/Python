*  **Python Magic methods** are the methods starting and ending with double underscores ‘__’. They are defined by built-in classes in Python and commonly used for **operator overloading**.

*  They are also called **Dunder methods**, Dunder here means **“Double Under (Underscores)”**.


````python
class A:
   def __init__(self, a):
       self.a = a

  #   #naveen
   def __str__(self):
        return f"str--->,{self.a}"

  #  def __repr__(self):
  #       return f"repr--->,{self.a}"


   def __lt__(self, other):
       if(self.a < other.a):
           return "ob1 is less than ob2"
       else:
           return f"{other.a} is less than {self.a}"

   def __eq__(self, other):
       if(self.a == other.a):
           return "Both are equal"
       else:
           return "Not equal"

   def __sub__(self, other):
        print("calling __sub__")
        return self.a + other.a


ob1 = A(2)
ob2 = A(3)
print([ob1,ob2])
print(ob1 < ob2)

object_1 = A('z')
object_2 = A('a')
print("sample",object_1 < object_2)


ob3 = A(4)
ob4 = A(4)
print(ob3 == ob4)

object_3 = A('4')
object_4 = A('4')
print(object_3 == object_4)

ob5 = A(34)
ob6 = A(66)
print(ob5.__sub__(ob6))

ob7 = A('34')
ob8 = A('66')
print(ob7.__sub__(ob8))

ob7 = A('34')
ob8 = A('66')
print(ob5 - ob6)
print(ob7 - ob8)
````

## MRO
<br>

`Method Resolution Order(MRO) of a class:`

* It is the order in which Python looks for a method in a hierarchy of classes
* To determine which method it actually calls

example,


````python
class A:
    def rk(self):
        print("In class A")

class B(A):
    def rk(self):
        print("In class B")

class C(A):
    def rk(self):
        print("In class C")

# classes ordering
class D(C,B):
    pass

r = D()
r.rk()
````

**Difference between Method Overloading and Method Overriding in Python**

<br>

**Method Overloading:**

*   is an example of Compile time polymorphism
*   more than one method of the same class shares the same method name having different signatures
*   there is no need of more than one class for method overloading.


**Method Overriding:**

 *  is an example of run time polymorphism.
 *  method that is already provided by the parent class is provided by the child class
 * It is used to change the behavior of existing methods and there is a need for at least two classes for method overriding


````python
# naveen #
class example:
   def add(self, a, b):
      x = a+b
      return x
   def add(self, a, b, c):
      x = a+b+c
      return x

obj = example()

print(dir(obj))

print (obj.add(10,20,30))
print (obj.add(10,20))
````

````python

''' Method Overloading '''
def add(datatype, *args):

    # if datatype is int
    # initialize answer as 0
    if datatype =='int':
        answer = 0

    # if datatype is str
    # initialize answer as ''
    if datatype =='str':
        answer =''

    # Traverse through the arguments
    for x in args:

        # This will do addition if the
        # arguments are int. Or concatenation
        # if the arguments are str
        answer = answer + x

    print(answer)

# Integer
add('int', 5, 6)

# String
add('str', 'Hi ', 'Geeks')
````

````python


''' Method Overriding'''
class A:

    def fun1(self):
        print('feature_1 of class A')

    def fun2(self):
        print('feature_2 of class A')


class B(A):

    # Modified function that is
    # already exist in class A
    def fun1(self):
        # print('Modified feature_1 of class A by class B')
        super().fun1()

    def fun3(self):
        print('feature_3 of class B')


# Create instance
obj = B()
# ob2 = A()

# Call the override function
obj.fun1()
````

##**Data Abstraction**



*   is a programming concept that hides complex implementation details while exposing only essential information and functionalities to users


````python
from abc import ABC, abstractmethod
class democlass(ABC):

   @abstractmethod
   def method1(self):
      print ("abstract method")
      return

   # @abstractmethod
   def method2(self):
      print ("concrete method")


class concreteclass(democlass):
   def method1(self):
      super().method1()
      return

obj = concreteclass()
obj.method1()
obj.method2()
````

  **property( )** is a built-in function for creating and returning a property object.



````python
# Python program to illustrate property() function
class Person:
    def __init__(self, name):
        self._name = name

    # getter function
    def get_name(self):
        print('Getting name...')
        return self._name

    # setter function
    def set_name(self, value):
        print('Setting name to:', value)
        self._name = value

    # deleter function
    def del_name(self):
        print('Deleting name...')
        del self._name

    # Set property() to use get_name, set_name and del_name methods
    # property(fget=None, fset=None, fdel=None, doc=None)
    name = property(get_name, set_name, del_name)

p = Person('David')
print(p.name)
p.name = 'Rocky'
del p.name
````

````python
# Python program to illustrate property() function
class Person1:
    def __init__(self, name):
        self._name = name

    # getter function
    def a_name(self):
        print('Getting name...')
        print(self._name)
        return self._name

    # setter function
    def b_name(self, value):
        print('\n')
        print('Setting name to:', value)
        self._name = value
        print(self._name)

    # deleter function
    def c_name(self):
        print('\n')
        print('Deleting name...')
        del self._name
        # print(self._name)

    # Set property() to use get_name, set_name and del_name methods
    name__update = property(a_name, b_name, c_name)


p = Person1('David')
print(p.name__update)
p.name__update = 'Rocky'
del p.name__update
````

### Class method vs Static Method
The difference between the Class method and the static method is:

* class method takes **cls** as the first parameter while a static method needs **no specific parameters**

* We use **@classmethod** decorator in python to create a class method and we use **@staticmethod** decorator to create a static method in python



````python
class MyClass:
    def __init__(self, value):
        self.value = value

    @staticmethod
    def get_max_value(x, y):
        return max(x, y)

    @classmethod
    def add(cls, x, y):

        return x + y

    def sub(cls, x, y):
        return x - y


````

## **Encapsulation**

*  Access through Methods

*  Control and Security

*  Data Hiding

<br>

1. Protected Members:

  * identified with a single underscore (_)
  * accessed only within the class or its subclasses.

2. Public Members:

  * accessible from anywhere, both inside and outside the class

3. Private Members:

  * identified with a double underscore (__)
  * cannot be accessed directly from outside the class.

````python
class Protected:
    def __init__(s):
        s._age = 30  # Protected attribute

class Subclass(Protected):
    def display_age(s):
        print(s._age)  # Accessible in subclass

obj = Subclass()
obj.display_age()

print(obj._age)
````

````python
class Public:
    def __init__(self):
        self.name = "John"  # Public attribute

    def display_name(self):
        print(self.name)  # Public method

obj = Public()
obj.display_name()  # Accessible
print(obj.name)  # Accessible
````

````python
class Private:
    __sal = 10000
    def __init__(self):
        self.__salary = 50000  # Private attribute

    def salary(self):
        return self.__salary  # Access through public method

obj = Private()

# _Private__salary

print(obj.salary())
# print(obj._Private__sal)
````

````python
class BankAccount:
    balance_value = 10
    __private_balance = 50
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        self.__log_transaction(amount, "deposit")  # call private method inside class

    def __log_transaction(self, amount, type_):
        print(f"Transaction: {type_} of {amount}")


account = BankAccount(1000)
account.deposit(500)
# print(account.balance_Value)
print(account._BankAccount__private_balance)
# account.__log_transaction(500, "deposit")
````

````python
class Demo:
    def __init__(self):
        self.a = 10        # Public
        self._b = 20       # Protected
        self.__c = 30      # Private

    def show_inside(self):
        print("Inside class:")
        print("Public:", self.a)
        print("Protected:", self._b)
        print("Private:", self.__c)


d = Demo()


print("Public outside:", d.a)


print("Protected outside:", d._b)

# print(d.__c)   # AttributeError

d.show_inside()

print("Private using name mangling:", d._Demo__c)
````
