# Decorators:
  * A decorator is a design pattern in Python that allows a user to add new functionality to an existing object without modifying its structure.
  * It is mostly used for login and authentication purpose
  * It is defined as @ at the begining of the function name
# Types:
    * Function Decorators
    * Class Decorators
    * Build-in decorators


# Function Decorators:
  * The most common type of decorator, which takes a function as input and returns a new function. The example belowdecorators.png demonstrates this type

```python
def my_decorator(func):
    def wrap():
        print("Before calling the function.")
        func()
        print("After calling the function.")
    return wrap

@my_decorator
def say_hello():
    print("Hello, World!")

say_hello()
```

# Class Decorator:
  * A class decorator in Python is similar to a function decorator, but it is used to modify or enhance the behavior of a class.
  * A class decorator takes a class as an argument and returns a new class, potentially modifying its behavior.

```python
def add_hello_method(cls):
    def hello(self):
      print("name---->",self.name)
      return "Hello from the decorator!"
    # cls.hello = hello
    cls.nae = "navis"
    return cls

@add_hello_method
class MyClass:
    def __init__(self, name):
        self.name = name

obj = MyClass("Alice")

# print(obj.hello())
print(obj.nae)
```

# Map Function in python:
  * The map() function is used to apply a given function to every item of an iterable, such as a list or tuple, and returns a map object
  * **Syntax:**
        map(function, iterables)

```python
# without map function
def displayName(name):
  print("hi!,", name)
  return True

names = ["arun", "akash", "sakthi"]
result = []
for name in names:
  result.append(displayName(name))
print(tuple(result))
```

```python
def displayName(name):
  print("hi!,",name)
  return True
names=["arun","akash","sakthi"]
result = map(displayName,names)
# result =map(lambda names:"hii"+names,names)
print(tuple(result))
```

# Filter Function:
  * The filter() method filters the given sequence with the help of a function that tests each element in the sequence to be true or not.
  * **syntax**
        filter(function, iterable)

```python
def displayName(name):
  return name=="arun"
names=["arun","akash","sakthi"]
result = filter(displayName,names)
print(list(result))
```

# Reduce function:
  * The reduce function is used to apply a function of two arguments cumulatively to the items of a list (or any iterable), from left to right, so as to reduce the iterable to a single value. Its not a method of the list but a function in the functools module.
  * **syntax:**
        reduce(function, sequence, initial=None)

```python
from functools import reduce

def add(x, y):
    return x + y

a = [1, 2, 3, 4, 5]
b=10
res = reduce(add, a, b)

print(res)
```

# Recursive function:
  * It means that a function calls itself. This has the benefit of meaning that you can loop through data to reach a result.
  * By using recursive, we donot need to call the function again and again

```python
def recursiveFunction(val):
  result = 0
  if val>0:
    result = val + recursiveFunction(val-1)                                      # 5+(4+(3+(2+(1+recursiveFunction(0))))
    print("result----->",result)
  return result
recursiveFunction(5)
```

# Generator Function in Python:
  * A generator function is a special type of function that returns an iterator object.
  * Instead of using return to send back a single value, generator functions use yield to produce a series of results over time.

```python
# def fun(max):
#     for i in range(0,max):
#         return i


# ctr = fun(5)
# print(type(ctr))
# for ctr in ctr:
#     print(ctr,"ctr---<")


def fun(i):
    yield i-1
    yield i
    yield i+1
ctr = fun(5)
for ctr in ctr:
    print(ctr, "ctr<-------")
```

```python
import sys

def normal():
  return [i for i in range(1000000)]

def generator():
  for i in range(1000000):
    yield i

print(sys.getsizeof(normal()))
print(sys.getsizeof(generator()))
```

# Lambda Function:
  * Lambda is an anonymous function.
  * Lambda is the keyword used for lambda function
  * A lambda function can take any number of arguments, but can only have one expression.
  * **Syntax**: lambda args:exp



```python
              ##### Without lambda function #####
def function(x):
  x = x+100
  return x
print("The result value is:",function(20))

            ##### With lambda function ######
x = lambda x:x+100
print("The result value is:",x(20))
```
